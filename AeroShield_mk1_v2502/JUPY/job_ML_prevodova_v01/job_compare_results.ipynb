```python
# imports and other setup

%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt
```

## Porovnanie nameraných prevodových charakteristík



```python
# -----------------------------------
# Dátové súbory:

data_file_path = './dataRepo/results/'




data_files = {
    'default': 'prevodovaCh_from_dataFile_2025_02_20_10_45_54_copy.txt.txt',
    'addedweight': 'prevodovaCh_from_dataFile_2025_02_20_14_26_13_copy.txt.txt',

}

```

```python
fig, ax = plt.subplots()

for label, file_name in data_files.items():
    file_path = data_file_path + file_name
    data = np.loadtxt(
        file_path, 
        delimiter=',',
        skiprows=0,
        )
    ax.plot(data[:, 0], data[:, 1], 
            label=label,
                # color='black',
                marker='o', 
                ms=5,
                # mec='black',
                mfc='None',
                lw=0,
            )


ax.set_xlabel('in [%]')
ax.set_ylabel('out [deg]')
ax.set_title('Porovnanie prevodových charakteristík')
ax.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
